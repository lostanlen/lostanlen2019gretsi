```python
import kymatio
import numpy as np
import scipy.signal
import torch
```

```python
N = 2**10
f0 = 16

def generate(fourier_decay, odd_to_even, f0=16, N=2**10):
    n_partials = int(N/(2*f0)) - 1
    t = np.linspace(0, 1, N, endpoint=False)
    partials = np.zeros((n_partials, N))
    for partial_id in range(n_partials):
        frequency = (1+partial_id) * f0
        if (partial_id % 2):
            amplitude = 1/(1+partial_id)**fourier_decay
        else:
            amplitude = odd_to_even/(1+partial_id)**fourier_decay
        partial = amplitude * np.cos(2*np.pi*frequency*t)
        partials[partial_id, :] = partial
    x = np.sum(partials, axis=0) * scipy.signal.hann(N)
    return x / np.max(x)#np.linalg.norm(x)

import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt


alphas = np.ravel(np.tile(np.array([0.5, 1.5, 2.5]), (1, 3)))
rs = np.ravel(np.tile(np.array([-3.0, -1.0, 1.0]), (3, 1)).T)
fig, axs = plt.subplots(3, 3, figsize=(12, 10), sharex=True, sharey=True)
axs = axs.flatten()

for i in range(len(fourier_decays)):
    alpha = alphas[i]
    r = rs[i]
    x = generate(alpha, 2**r, f0=f0, N=N)
    axs[i].plot(x)
    axs[i].set_xlim(0, N)
    axs[i].set_ylim(-1.1, 1.1)
    axs[i].set_title("α = {:.2f} ; r = {:.2f}".format(alpha, r))
    axs[i].set_xticks(np.linspace(0, N, 5))
    axs[i].set_xticklabels(np.linspace(-0.5, 0.5, 5))
    axs[i].grid('on', linestyle='--', alpha=0.5)
    axs[i].set_xlabel("Time")
    axs[i].set_ylabel("Amplitude")
```

```python
import tqdm

def build_dataset(min_alpha, max_alpha, min_r, max_r, n_alpha=100, n_r=100):
    r_range = np.linspace(min_r, max_r, n_r, endpoint=True)
    alpha_range = np.linspace(min_alpha, max_alpha, n_alpha, endpoint=True)
    N = 2**10
    X = np.zeros((n_r, n_alpha, N))
    for i, r in tqdm.tqdm(enumerate(r_range)):
        for j, alpha in enumerate(alpha_range):
            X[i, j, :] = generate(alpha, 2**r, f0=16, N=N)
            
X = build_dataset(0.0, 4.0, -3.0, 1.0)
```
